```python
import os
import torch
import torch.nn as nn
import numpy as np
import random
```

```python
def get_batch_from_data(dataset, batch_size, context_len, device):
    sts = [random.randint(0,len(dataset)-context_len-2) for _ in range(batch_size)]
    data_load = [dataset[st:st+context_len] for st in sts]
    target = [dataset[st+1:st+context_len+1] for st in sts]
    return torch.tensor(data_load,device=device),torch.tensor(target,device=device)
    
```
